```sql
-- 1. Create a stored procedure to display the restaurant name, type and cuisine where the
-- table booking is not zero.
CREATE OR ALTER PROCEDURE sp_restaurant_with_table_booking
AS
BEGIN
    SELECT RestaurantName,RestaurantType,CuisinesType FROM restaurants 
    GROUP BY RestaurantName,RestaurantType,CuisinesType,TableBooking
    HAVING TableBooking='Yes' AND COUNT(TableBooking)>0
END
```

```sql
EXEC sp_restaurant_with_table_booking
```

```sql
-- 2. Create a transaction and update the cuisine type Cafe to Cafeteria. Check the result
-- and rollback it.
BEGIN TRANSACTION;

UPDATE restaurants
SET RestaurantType = 'Cafeteria'
WHERE RestaurantType = 'Cafe';

SELECT RestaurantName,RestaurantType,CuisinesType FROM restaurants;

ROLLBACK TRANSACTION;
```

```sql
-- 3. Generate a row number column and find the top 5 areas with the highest rating of
-- restaurants.
WITH ctx as(
SELECT 
ROW_NUMBER() OVER(order by AVG(Rating) desc) as RowNumber,
Area, 
AVG(Rating) as AvgRating
FROM restaurants
GROUP BY Area
)

SELECT * FROM ctx WHERE RowNumber<=5
```

```sql
-- 5. Write a query to Create a Top rating view to store the generated top 5 highest rating of
-- restaurants.
SELECT ROW_NUMBER() OVER(order by Rating desc) as Rank, RestaurantName, Rating
FROM restaurants
```
