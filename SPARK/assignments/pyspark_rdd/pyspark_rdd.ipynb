```python
import findspark
findspark.find()
findspark.init()
```

```python
from pyspark import SparkContext

sc = SparkContext(appName='WordCount', master='local[*]')
sc
```

```python
text = sc.textFile('./RomeoJuliet.txt')
text.take(10)
```

```python
rdd=text.flatMap(lambda line: line.split(' '))\
    .map(lambda word: (word, 1))\
        .reduceByKey(lambda v1,v2:v1+v2)
```

```python
for word in rdd.collect():
    print(word)
```

```python
sc.stop()
```
