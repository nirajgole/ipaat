The methods should return a new pandas DataFrame with keys as column

````python
def create_data_frame(keys, values):
    return pd.DataFrame(data=values, columns=keys)


create_data_frame(["One", "Two"], [["X", "Y"], ["A", "B"]])
````

````python
def concat_data_frame(d1, d2):

    return pd.concat([d1, d2], ignore_index=True)


df1 = pd.DataFrame([[1, 2, 3], [1, 2, 3]])
df2 = pd.DataFrame([[1, 2, 3], [1, 2, 3]])
print(concat_data_frame(df1, df2))
````

```
5. Write a method that will take a column name as argument and return the
name of the column with which the given column has the highest
correlation.
The data to be used is the cars dataset.
The returned value should not the column named that was passed as the
parameters, e.g. : get_max_correlated_column('mpg') -> should return 'drat

```

````python
def get_max_correlated_column(df1,col):
    """
    Get the column with maximum correlation to mpg.
    """    
    max_corr=0
    col1=None
    df2=df1.iloc[:,1:]
    for i in df2:
        if i!='mpg':
            c=df2[col].corr(df2[i])
            if max_corr<c:
                max_corr=c
                col1=i
    return col1
    
get_max_correlated_column(df,'mpg')
````
